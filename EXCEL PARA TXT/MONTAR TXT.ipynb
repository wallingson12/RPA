```python
import pandas as pd
from datetime import datetime

# Ler o arquivo Excel
df = pd.read_excel('teste.xlsx')

# Obter os dados da coluna CNPJ do Tomador
cnpj_tomador = str(df['CNPJ do Tomador'].iloc[0])

# Obter os dados do nome do Tomador
nome_tomador = str(df['Nome do Tomador'].iloc[0])

# Obter os dados do endereço
endereco = str(df['Endereço'].iloc[0])

# Obter os dados do bairro
bairro = str(df['Bairro'].iloc[0])

# Obter os dados da cidade do prestador
cidade_prestador = str(df['Cidade do Prestador'].iloc[0])

# Obter os dados do CNPJ do Prestador
cnpj_prestador = str(df['CNPJ do Prestador'].iloc[0])

# Obter os dados do nome do Prestador
nome_prestador = str(df['Nome do Prestador'].iloc[0])

# Obter os dados da Discriminação dos Serviços
discriminacao_servicos = str(df['Discriminação dos Serviços'].iloc[0])

# Remover "." e "," do valor da nota
valor_nota = str(df['Valor Total'].iloc[0]).replace('.', '').replace(',', '')

# Lista com os valores desejados
inscrição_municipal = ["00035564"]

data_emissao_str = str(df['Data de Emissão'].iloc[0])

# Converter a data de emissão para um objeto de data
data_emissao = datetime.strptime(data_emissao_str, '%d/%m/%Y').date()

# Converter a data de emissão para o formato AAAAMMDD
data_inicio_mes = data_emissao.strftime('%Y%m') + '01'
ultimo_dia_mes = (data_emissao.replace(day=1, month=data_emissao.month+1) - pd.DateOffset(days=1)).strftime('%Y%m%d')

# Criar o conteúdo do cabeçalho
cabecalho = (
    f"{str(10).zfill(2)}" +
    f"{str(3).zfill(3)}" +
    "2" +
    cnpj_tomador.ljust(8, '0') +
    "0" *7 +
    " ".join(inscrição_municipal) +
    data_inicio_mes +
    ultimo_dia_mes
)

# Definir o nome do arquivo de saída
nome_arquivo_saida = 'testeFINAL.txt'

# Escrever o cabeçalho no arquivo de saída
with open(nome_arquivo_saida, 'w') as arquivo:
    arquivo.write(cabecalho + '\n')

# Criar o conteúdo da linha 2
linha2 = (
    f"{str(40).zfill(2)} " +
    "1" +
    " " * 4 +
    "1"
    +
    str(df['Número da Nota'].iloc[0]).zfill(14) +
    data_emissao.strftime('%Y%m%d') +
    "1" +
    "2" +
    cnpj_prestador.ljust(14) +
    " " * 67 +
    nome_prestador.ljust(112) +
    "XXX" +
    " " * 50 +
    endereco.ljust(125) +
    " " * 23 +
    bairro.ljust(48) +
    cidade_prestador.ljust(109) +
    "02" +
    " " * 54 +
    "0"+
    str(df['Código de Serviço'].iloc[0]).zfill(5) +
    " " * 11 +
    str(df['Código de Serviço'].iloc[0]).zfill(5) +
    str(df['Alíquota'].iloc[0]).zfill(2) +
    "0" *16 +
    valor_nota.zfill(4) +
    "0" * 14 +
    " " *30 +
    "0" *16 +
    data_emissao.strftime('%Y%m%d') +
    " " * 30 +
    discriminacao_servicos.ljust(30)
)

# Adicionar a linha 2 no arquivo de saída
with open(nome_arquivo_saida, 'a') as arquivo:
    arquivo.write(linha2 + '\n')

# Calcular o número de notas
numero_notas = len(df)

# Calcular o total do valor das notas
total_valor_notas = df['Valor Total'].sum()

# Definir os valores das deduções e descontos
valor_deducoes = 0
valor_descontos = 0
valor_descontos_incond = 0

# Criar o conteúdo do rodapé
rodape = (
    f"{str(90).zfill(2)}" +
    str(numero_notas).zfill(8) +
    str(total_valor_notas).replace('.', '').replace(',', '').zfill(15) +
    str(valor_deducoes).replace('.', '').replace(',', '').zfill(15) +
    str(valor_descontos).replace('.', '').replace(',', '').zfill(15) +
    str(valor_descontos_incond).replace('.', '').replace(',', '').zfill(15)+
    "0"
)

# Adicionar o rodapé no arquivo de saída
with open(nome_arquivo_saida, 'a') as arquivo:
    arquivo.write(rodape + '\n')

print(f"Arquivo '{nome_arquivo_saida}' criado com sucesso!")
```
